```python
import pandas as pd
import glob

#Use glob to find all Excel files in the Colab environment
xls_files = glob.glob("*.xls")

print(f"Found {len(xls_files)}files.")

all_data = []

for file in xls_files:
  try:
    df = pd.read_excel(file, skiprows=15)
    df['Source_File'] = file
    all_data.append(df)
    print(f"Loaded: {file}")
  except Exception as e:
    print(f"Error Loading {file}: {e}")

  #Combine into single dataframe
  df_all = pd.concat(all_data,ignore_index = True)

  #Preview
  df_all.head()
```

```python
#To track which month/year each row came from:

import re
def extract_month_year(filename):
  match = re.search(r'(january|february|march|april|may|june|july|august|september|october|november|december)[-_]?\d{4}',
                    filename,re.IGNORECASE)
  return match.group(0).title() if match else "Unknown"
df_all['Month_Year'] = df_all['Source_File'].apply(extract_month_year)
df_all.head()
```
